***Question 1:*** What is Computational Linguistics and how does it relate to NLP?

***Answer:***

Computational Linguistics is the scientific study of language using
computational models, focusing on formal theories of grammar and
syntax analyzed via algorithms. It relates to NLP as its theoretical
foundation—NLP applies these models practically for tasks like machine
translation or chatbots, bridging linguistics with engineering.

***Question 2:*** Briefly describe the historical evolution of Natural Language Processing.

***Answer:***

NLP began in the 1950s with rule-based systems like the Georgetown-IBM
experiment (1954) for Russian-English translation. The 1970s-80s shifted
to statistical methods (e.g., Hidden Markov Models), 1990s introduced
machine learning, and post-2010 deep learning (RNNs, Transformers)
enabled modern applications like GPT models.


***Question 3:*** List and explain three major use cases of NLP in today’s tech industry.

***Answer:***

Key use cases include sentiment analysis for customer feedback (e.g.,
analyzing reviews), machine translation (e.g., Google Translate), and
chatbots/virtual assistants (e.g., Siri for voice commands). These drive
industries like e-commerce, healthcare, and customer service.

***Question 4:*** What is text normalization and why is it essential in text processing tasks?

***Answer:***

Text normalization standardizes raw text by converting to lowercase,
removing punctuation/noise, expanding contractions, and handling
accents—e.g., "Don't", "café" becomes "dont cafe". It's essential to reduce
vocabulary size, improve model consistency, and boost accuracy in tasks
like classification.

***Question 5:*** Compare and contrast stemming and lemmatization with suitable
examples.

***Answer:***

Stemming crudely chops words to roots (e.g., "running" → "run",
"stripes" → "strip") using rules like Porter Stemmer, which can produce
non-words. Lemmatization uses dictionaries/POS tags for valid base
forms (e.g., "running" → "run", "stripes" → "stripe"), making it more
accurate but slower.

```python
'''
Question 7: Given the sample paragraph below, perform string tokenization and
frequency distribution using Python and NLTK:
“Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.”
(Include your Python code and output in the code box below.)
Answer:
'''
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
nltk.download('punkt', quiet=True)
paragraph = "Natural language processing enables computers to understand human language. It enables machines to understand, interpret, and generate human language. Applications of NLP include chatbots, sentiment analysis, and machine translation. As technology advances, the role of NLP in modern solutions is becoming increasingly critical."
tokens = word_tokenize(paragraph.lower())
freq_dist = FreqDist(tokens)
print("Tokens (first 20):", tokens[:20])
print("Top 10 Frequency Distribution:")
print(freq_dist.most_common(10))

#OUTPUT:-
Tokens (first 20): ['natural', 'language', 'processing', 'enables', 'computers', 'to',
'understand', 'human', 'language', '.', 'it', 'enables', 'machines', 'to', 'understand', ',',
'interpret', ',', 'and', 'generate']
Top 10 Frequency Distribution:
[(',', 5), ('.', 4), ('language', 3), ('enables', 2), ('to', 2), ('understand', 2), ('human', 2), ('and',
2), ('of', 2), ('nlp', 2)]
```

```python
'''
Question 8: Create a custom annotator using spaCy or NLTK that identifies and labels
proper nouns in a given text.
(Include your Python code and output in the code box below.)
Answer:

'''

import spacy
nlp = spacy.load("en_core_web_sm")
text = "Apple is a company based in Cupertino. Elon Musk founded Tesla in 2003."
doc = nlp(text)
print("Proper Nouns (Entities):")
for ent in doc.ents:
  if ent.label_ in ['PERSON', 'ORG', 'GPE']:
    print(f"{ent.text} -> {ent.label_}")
```

```python
# Question 9: Using Genism, demonstrate how to train a simple Word2Vec model on the following dataset consisting of example sentences:
# dataset = [
#  "Natural language processing enables computers to understand human language",
#  "Word embeddings are a type of word representation that allows words with similar meaning to have similar representation",
#  "Word2Vec is a popular word embedding technique used in many NLP applications",  "Text preprocessing is a critical step before training word embeddings",  "Tokenization and normalization help clean raw text for modeling"
# ]
# Write code that tokenizes the dataset, preprocesses it, and trains a Word2Vec model using Gensim.
# (Include your Python code and output in the code box below.)
# Answer:

from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt', quiet=True)

dataset = [
    "Natural language processing enables computers to understand human language",
    "Word embeddings are a type of word representation that allows words with similar meaning to have similar representation",
    "Word2Vec is a popular word embedding technique used in many NLP applications",
    "Text preprocessing is a critical step before training word embeddings",
    "Tokenization and normalization help clean raw text for modeling"
]

tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in dataset]
model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)
print("Vocabulary size:", len(model.wv))
print("Vector for 'language' (first 5 dims):", model.wv['language'][:5])


#OUTPUT:
Vocabulary size: 46
Vector for 'language' (first 5 dims): [-0.00958061  0.00894419  0.00416531  0.00923353  0.00664613]
```

```python
# Question 10: Imagine you are a data scientist at a fintech startup. You’ve been tasked with analyzing customer feedback. Outline the steps you would take to clean, process, and extract useful insights using NLP techniques from thousands of customer reviews.
# (Include your Python code and output in the code box below.)
# Answer:
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)

reviews = ["Great app, fast transfers!", "Poor customer service, slow response.", "Love the security features."]
stop_words = set(stopwords.words('english'))

def process_review(review):
    tokens = word_tokenize(review.lower())
    filtered = [w for w in tokens if w.isalnum() and w not in stop_words]
    blob = TextBlob(' '.join(filtered))
    return blob.sentiment.polarity, filtered

insights = [process_review(r) for r in reviews]
print("Sentiment polarities:", [s[0] for s in insights])
print("Sample keywords from first review:", insights[0][1])

#OUTPUT:
Sentiment polarities: [0.5, -0.35000000000000003, 0.5]
Sample keywords from first review: ['great', 'app', 'fast', 'transfers']
```
